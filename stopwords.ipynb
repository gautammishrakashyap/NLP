```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download required datasets
nltk.download('punkt')
nltk.download('stopwords')

# Initialize stemmer
stemmer = PorterStemmer()

# Example sentences
sentence = ["This is a simple example sentence.", "It contains stopwords and needs stemming."]

# Process each sentence
for i in range(len(sentence)):
    words = word_tokenize(sentence[i])  # Tokenize words
    words = [stemmer.stem(word) for word in words if word.lower() not in set(stopwords.words('english'))]  
    sentence[i] = ' '.join(words)  # Join words back into sentence

print(sentence)
```
